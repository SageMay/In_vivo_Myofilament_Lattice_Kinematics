# Purpose

View unprocessed data and verify that the peak detection in process script worked (for acceptable trials) by comparing binary peak times with the raw EMG signal. Acceptable trials are trials where there was 
 
    1) Trackable x-ray diffraction data 
    2) Continuous flight throughout trial 
    3) an EMG whose DLM peaks could reliably be identified
    
The EMG signal often shows traces of both the DLM, quickly followed by the DVM muscle's activation. They can be disambiguated because the DLM peaks are larger due to the anatomical location where electrodes were inserted. 

```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
from Process import process, Preprocess_EMG
import scipy.io
import glob

DAQ_hertz = 25000
```

```python
all_trials = ['m07_t08_15', 'm10_t03_16', 'm10_t04_16', 'm12_t01_16', 'm14_t04_16','m14_t06_16', 'm15_t02_16','m07_t01_15', 'm07_t03_15', 'm07_t06_15','m10_t02_16','m11_t02_16','m11_t04_16','m12_t02_16','m14_t05_16', 'm14_t03_16', 'm15_t01_16', 'm15_t03_16'] 

acceptable_trials = ['m07_t01_15', 'm07_t03_15', 'm07_t06_15','m10_t02_16','m11_t02_16','m11_t04_16','m12_t02_16','m14_t05_16', 'm14_t03_16', 'm15_t01_16', 'm15_t03_16']

for t in acceptable_trials:
    EMG_f = '/Users/sagemalingen/Desktop/A17/Raw_Data/A17_RawData/DAQ_files_Argonnedec2017/' + t + '.mat'
    proc_f = 'ProcessedData/' + t + '_det.csv'
    
    # Import raw EMG and extract signal from the .mat file and put it into a dataframe; also drop the periods before and after the trigger.
    # The trigger is recorded in channel 0; it is 0 while the beam shutter was closed and 1 when the beam shutter was open. 
    EMG = pandas.DataFrame()
    try:
        EMG = scipy.io.loadmat(EMG_f)
        EMG = EMG['Indata']
    except:
        print('EMG import failed')
         
    Trig = EMG[:,0]
    EMG_amplitude = EMG[:,2]

    EMG = pandas.DataFrame(np.arange(0,len(EMG)))*(1/DAQ_hertz)
    EMG['Trig'] = Trig
    EMG['EMG_amplitude'] = EMG_amplitude
    EMG = EMG.rename(columns = {0:'seconds'})
    EMG = EMG.where(EMG.Trig > .5).dropna(how = 'all')
    EMG = EMG.reset_index(drop = 'True')
    EMG['seconds'] = EMG.seconds - 1
    
    if EMG.EMG_amplitude.max() < (-1*EMG.EMG_amplitude).max():
        EMG.EMG_amplitude = -1*EMG.EMG_amplitude
    
    # Import processed file
    proc = pandas.read_csv(proc_f)
    # Plot all of the relevant data
    proc = proc[['D10', 'peaks', 'M3_c', 'M6_c', 'A59_c', 'M3_i', 'I20_I10', 'seconds']]
    proc.plot(subplots = 'True', figsize = (10,10))
    print(t)
    plt.show()
    plt.cla()
    
    # Plot red vertical lines when the peaks occured in the processed dataframe over the raw EMG signal to verify that peak identification was successful. 
    proc = proc.where(proc.peaks > .5).dropna(how = 'all')
    plt.plot(EMG.seconds, EMG.EMG_amplitude, c = 'black')
    for i in proc.seconds:
        plt.axvline(x=i, c = 'red', alpha = .5)
    plt.title(t)
    plt.show()
    
    
    
```
