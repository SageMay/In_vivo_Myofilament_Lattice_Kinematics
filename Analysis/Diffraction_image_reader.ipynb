# Purpose
Load raw tif images and convert them to a visualizable format.

```python
from matplotlib import pyplot as plt
%matplotlib inline
from matplotlib import cm
import imageio
from PIL import Image
from numpy import array
import numpy as np
import os
from numpy import linalg as LA
import scipy.io
import matplotlib.colors as colors
```

```python
# set up filenames

pathin = "/Users/sage/Desktop/ImagesIn/m15_t01_16_qf/"
pathout = "/Users/sage/Desktop/ImagesOut/m15_t01_16_qf_out/"
basename = "m15_t01_16_00"
mycmap = 'magma' # 'afmhot' and 'jet' are also nice color maps to use
images = []
folded = '_folded'
```

```python
for j in range(8,190):
    plt.clf()
    str2 = str(j)
    
    if j< 100:
        str2 = "0" + str(j)
    if j < 10:
        str2 = "00" + str(j)
    filenamein = pathin + basename + str2 + folded + '.tif'
    filenameout = pathout + str2 + folded + '.tif'
    im2 = imageio.imread(filenamein)
    remapped = np.log(30 + 3*im2)
    
    plt.imshow(remapped,cmap=mycmap)
    plt.savefig(filenameout)
    images.append(imageio.imread(filenameout))
    
    # delete individual plot so only the movie remains (optional)
    #os.remove(filenameout)
```

```python
# Use this if you want to save a gif ... 
imageio.mimsave(pathout + 'diffraction_movie.gif', images)
```
